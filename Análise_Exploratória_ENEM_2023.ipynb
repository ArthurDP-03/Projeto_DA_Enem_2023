###**Leitura dos Dados:**





- Ler arquivo csv do enem

```python

dados_enem = pd.read_csv('MICRODADOS_ENEM_2023.csv', sep = ';' ,encoding='latin-1')
dados_enem.head()
```

```python
dados_enem.shape
```

\* O número de inscrições do enem de 2023 teve um decrescimo de aproximadamente 1.100.000 de pessoas que se inscreveram em relação ao enem de 2019



```python
dados_enem.columns
```

```python
dados_enem_novo = dados_enem[['TP_SEXO', 'TP_FAIXA_ETARIA', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ESCOLA', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'TP_STATUS_REDACAO', 'NU_NOTA_REDACAO']]
dados_enem_novo
```

```python
# Removendo as pessoas que não realizaram todas as provas e anularam a redação
dados_enem_novo = dados_enem_novo[dados_enem_novo[['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT']].eq(1).all(axis=1)]
dados_enem_novo = dados_enem_novo.dropna(subset=['TP_STATUS_REDACAO'])

# Resetando o índice após a remoção
dados_enem_novo = dados_enem_novo.reset_index(drop=True)

# Exibindo o resultado
dados_enem_novo
```

colunas a serem expLoradas para anélise: 'TP_SEXO', 'TP_FAIXA_ETARIA' 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO', 'TP_ESCOLA', 'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3', 'NU_NOTA_COMP4',
'NU_NOTA_COMP5', 'NU_NOTA_REDACAO',

```python
dados_enem_novo = dados_enem_novo.dropna().reset_index(drop=True)
dados_enem_novo.head()
```

```python
dados_enem_novo
```

```python
#Padronizando as notas
from sklearn.preprocessing import StandardScaler

df_enem = pd.DataFrame(dados_enem_novo)

colunas_para_padronizar = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']

scaler = StandardScaler()

df_enem[colunas_para_padronizar] = scaler.fit_transform(df_enem[colunas_para_padronizar])

df_enem
```

```python
df_enem['TP_SEXO'] = df_enem['TP_SEXO'].apply(lambda sex: 1 if sex == 'M' else 0)
df_enem
```
